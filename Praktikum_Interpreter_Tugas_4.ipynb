```
# n = int(input("Masukkan ukuran : "))
n = 7

def square(n):
    for i in range(n):
        for j in range(n):
            print("x", end="")
        print()
square(n)
```

```
def squareEmpty(n):
    for i in range(n):
        for j in range(n):
            if i == 0 or i == n-1:
                print("x", end="")
            elif i != 0 and j != n-1 and j != 0:
                print(" ", end="")
            else:
                print("x", end="")
        print()
squareEmpty(n)
```

```
def rightTriangle(n):
    for i in range(1, n+1):
        for j in range(1, n+1):
            if i >= j:
                print("x", end="")
            else:
                print(" ", end="")
        print()
rightTriangle(n)
```

```
def rightTriangleEmpty(n):
    for i in range(n):
        for j in range(n):
            if j == 0 or j == i or i == n-1:
                print("x", end="")
            else:
                print(" ", end="")
        print()
rightTriangleEmpty(n)
```

```
def leftTriangle(n):
    for i in range(1, n+1):
        for j in range(1, n+1):
            if i >= n-j+1:
                print("x", end="")
            else:
                print(" ", end="")
        print()
leftTriangle(n)
```

```
def nNormal(n):
    for i in range(n):
        for j in range(n):
            if j == 0 or j == n-1 or j == i:
                print("x", end="")
            else:
                print(" ", end="")
        print()
nNormal(n)
```

```
def zNormal(n):
    for i in range(n):
        for j in range(n):
            if i == 0 or i == n-1 or j == n-i-1:
                print("x", end="")
            else:
                print(" ", end="")
        print()
zNormal(n)
```

```
def zFlipped(n):
    for i in range(n):
        for j in range(n):
            if i == 0 or i == n-1 or j == i:
                print("x", end="")
            else:
                print(" ", end="")
        print()
zFlipped(n)
```

```
def xCross(n):
    for i in range(n):
        for j in range(n):
            if j == i or j == n-i-1:
                print("x", end="")
            else:
                print(" ", end="")
        print()
xCross(n)
```
